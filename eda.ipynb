# Exploratoy Data Analysis

this dataset describes how ai affects students

Initializing Data Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

importing dataset

```python
data = pd.read_csv('data.csv')
```

dataset:

```python
data
```

```python
print("no of rows:", data.shape[0])
print("no of columns:", data.shape[1])
print("")
for i in data.columns:
    print(f"column name: {i},\n data type: {data[i].dtype}, \n no of unique values: {data[i].nunique()}")
```

describtion of the data

```python
print("no of missing values:", data.isnull().sum().sum())
print("no of duplicate rows:", data.duplicated().sum())
print("")
print("data types:\n", data.dtypes)
print("")
print("descriptive statistics:\n", data.describe())
```

```python
print("this shows the distribution of the Prompt Engineering Skill column in the dataset. The bar chart represents the frequency of each unique value in the column, with the percentage of each value displayed on the chart.")
promptengineering=data['Prompt_Engineering_Skill']
plt.bar(promptengineering.value_counts().index, promptengineering.value_counts(), label='Frequency')
plt.xlabel('Prompt Engineering Skill')
plt.ylabel('Frequency')
plt.title('Distribution of Prompt Engineering Skill')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
print("this shows the distribution of the Primary Use Case column in the dataset. The pie chart represents the proportion of each unique value in the column, with the percentage of each value displayed on the chart.")
primary_use_case=data['Primary_Use_Case']
plt.pie(primary_use_case.value_counts(), labels=primary_use_case.value_counts().index, autopct='%1.1f%%')
```

```python
print("this show out of all the students in the dataset, how many are in each year of study. The bar chart represents the frequency of each unique value in the column, with the percentage of each value displayed on the chart.")
yr_of_study=data['Year_of_Study']
plt.bar(yr_of_study.value_counts().index, yr_of_study.value_counts(), label='Frequency')
```

```python
print("this shows out of all the students in the datasets how many people use how many no of AIs do they use")
diversity=data['Tool_Diversity']
plt.barh(diversity.value_counts().index, diversity.value_counts(), label='Frequency')
```
